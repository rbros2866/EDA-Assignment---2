```python
# Interquartile Range (IQR)
Q1 = flight_data['Price'].quantile(0.25)
Q3 = flight_data['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_count = ((flight_data['Price']< lower_bound) | (flight_data['Price'] > upper_bound)).sum()

print(f"Number of outliers: {outliers_count}")
```

```python
# Function to convert time in minutes
def time_to_minutes(time_str):
    parts = time_str.split()
    hours = 0
    minutes = 0
    for part in parts:
        if 'h' in part:
            hours = int(part.replace('h', ''))
        elif 'm' in part:
            minutes = int(part.replace('m', ''))
    return hours * 60 + minutes if hours or minutes else None


# Extract hours and minutes from 'Flight_Duration' and convert to minutes
flight_data['Duration_in_minutes'] = flight_data['Duration'].apply(lambda x: time_to_minutes(x))
```

```python
# Create a scatter plot with hue based on 'Airline' using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=flight_data, x='Duration_in_minutes', y='Price', hue='Airline', palette='bright', alpha=0.7)
plt.xlabel('Flight Duration (minutes)')
plt.ylabel('Price')
plt.title('Flight Duration vs Price')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
missing_values = apps_data.isnull().sum()

print(missing_values)
```

**Missing values, represented as NaN or None, can impact analysis in various ways:**

**Biased Analysis:** Missing values can lead to biased analysis if not handled properly, as some operations might exclude incomplete rows or columns.

**Reduced Sample Size:** Depending on the extent of missing data, your sample size for analysis may reduce, affecting the statistical power and generalizability of results.

**Inaccurate Insights:** Analysis involving missing values might provide inaccurate insights due to incomplete information or skewed distributions.

**Impact on Imputation:** If imputation methods are used to fill missing values, the chosen method can introduce additional uncertainty or bias into the dataset.

```python
# Fill missing 'Rating' values with the mean
apps_data['Rating'].fillna(apps_data['Rating'].mean(), inplace=True)

# Fill missing categorical values with the most frequent category
columns_with_missing = ['Type', 'Content Rating', 'Current Ver', 'Android Ver']
for col in columns_with_missing:
    apps_data[col].fillna(apps_data[col].mode()[0], inplace=True)
```

**Q12.** What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.

```python
# Function to convert 'Size' values to numeric
def convert_size_to_numeric(size):
    if 'M' in size:
        return float(size.replace('M', '')) * 1000  # Convert megabytes to KB
    elif 'k' in size:
        return float(size.replace('k', '')) * 1  
    else:
        return None  # For values without 'M' or 'k'

# Apply the conversion function to the 'Size' column
apps_data['Size'] = apps_data['Size'].apply(convert_size_to_numeric)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8)) 
ax = sns.scatterplot(x='Size', y='Rating', hue='Category', data=apps_data, alpha=0.7)
plt.xlabel('App Size in KB')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating with App Category')
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.yticks([i / 10 for i in range(0, 51)], labels=[f'{i / 10:.1f}' for i in range(0, 51)])  # Set y-axis ticks and labels

# Increase y-axis size
ax.set_ylim(-0.1, 5.1)  # Set the limits of the y-axis

plt.show()

```
